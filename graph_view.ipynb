```python
import glob
import matplotlib.pyplot as plt
import pandas as pd
from tqdm.notebook import trange
```

```python
train_raw = None

for p in glob.glob('train/*.csv'):
    data = pd.read_csv(p)
    if train_raw is None:
        train_raw = data
        continue
train_raw = train_raw.sort_values(['timestamp'])
```

```python
plt.plot(range(0, len(train_raw)), train_raw.C03)
```

```python
train_raw.drop(['timestamp'], axis=1, inplace=True)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=22, figsize=(70,9))
plt.axis('off')
for i in trange(0, len(train_raw.columns)):
    row = int(i / 22)
    col = i % 22
    axes[row, col].plot(range(0, len(train_raw[train_raw.columns[i]])), train_raw[train_raw.columns[i]].to_numpy())
    axes[row, col].set_title(train_raw.columns[i])
    axes[row, col].axis('off')
plt.show()
```
